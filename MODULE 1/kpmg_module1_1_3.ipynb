```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('/content/Transactions_1.0.xlsx')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Ensuring transaction id is unique
pd.Series(data['transaction_id']).is_unique
```

```python
# Make sure the product_id only contains legitimate digits
data['product_id'] = data['product_id'].astype('int64')
```

```python
# Sorting columns by customer_id and transaction_date
data.sort_values(['customer_id', 'transaction_date'])
```

```python
data['online_order'].value_counts()
```

```python
# Filling missing values using backward fill method
data['online_order'] = data['online_order'].fillna(method='bfill')
```

```python
data['brand'].value_counts()
```

```python
# Filling missing values using forward fill method
data['brand'] = data['brand'].fillna(method='ffill')
```

```python
data['product_line'].value_counts()
```

```python
# Filling missing values using forward fill method
data['product_line'] = data['product_line'].fillna(method='ffill')
```

```python
data['product_class'].value_counts()
```

```python
# Filling missing values using forward fill method
data['product_class'] = data['product_class'].fillna(method='ffill')
```

```python
data['product_size'].value_counts()
```

```python
# Filling missing values using forward fill method
data['product_size'] = data['product_size'].fillna(method='ffill')
```

```python
# work out the average cost upto 2 decimal places for each customer
pd.options.display.float_format = '{:,.2f}'.format
```

```python
customerStandardCost = data.groupby(['customer_id']).mean().round(decimals=2)
customerStandardCost
```

```python
# filling the missing data with the average costs for that customer
data['standard_cost'] = data['standard_cost'].fillna(customerStandardCost.standard_cost)
```

```python
# filling the missing data with the average first sold date
data['product_first_sold_date'] = data['product_first_sold_date'].fillna(data['product_first_sold_date'].mean())
```

```python
data.info()
```

```python
# print to csv after cleaning
data.to_csv('./Transaction_1.0_Cleaned.csv', index=False)
```
